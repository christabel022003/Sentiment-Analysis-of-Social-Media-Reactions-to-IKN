```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/My Drive/dataset ikn/ikn1-7_clean.csv'
df = pd.read_csv(file_path)

input_col = 'full_text'
df.head()
```

### **hapus karakter spesial**

```python
import re

def remove_special_characters(text):
    text = re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", str(text))
    text = re.sub(r"\d+", "", text)
    return text

df['cleaned_text'] = df[input_col].apply(remove_special_characters)
df[['cleaned_text']].head()
```

### **case folding**

```python
def case_folding(text):
    return text.lower()

df['case_folded_text'] = df['cleaned_text'].apply(case_folding)
df[['case_folded_text']].head()
```

### **Tokenisasi**

```python
!pip install nltk
```

```python
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')

def tokenize_text(text):
    return word_tokenize(text)

df['tokenized_text'] = df['case_folded_text'].apply(tokenize_text)
df[['tokenized_text']].head()
```

### **Stopwords**

```python
!pip install stopwords
```

```python
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('indonesian'))

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

df['stopwords_removed'] = df['tokenized_text'].apply(remove_stopwords)
df[['stopwords_removed']].head()
```

### **Stemming**

```python
!pip install sastrawi
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(tokens):
    return [stemmer.stem(word) for word in tokens]

df['stemmed_text'] = df['normalized_text'].apply(stemming)
df['final_text'] = df['stemmed_text'].apply(lambda x: ' '.join(x))
df[['final_text']].head()
```

```python
import pandas as pd

# Load dataset
file_path = '/content/drive/MyDrive/dataset ikn/ikn1-7_preprocessing.csv'
df = pd.read_csv(file_path)

assert 'final_text' in df.columns, "Kolom 'final_text' tidak ditemukan dalam dataset"

print(f"Dataset loaded. Total data: {len(df)}")
```

```python
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis", model="w11wo/indonesian-roberta-base-sentiment-classifier")

# Terapkan prediksi sentimen
def label_sentiment(text):
    result = sentiment_pipeline(str(text))
    # Updated label_map to match the model's output
    label_map = {'negative': 0, 'neutral': 1, 'positive': 2}  # Negatif: 0, Netral: 1, Positif: 2
    return label_map[result[0]['label']]

df['sentiment_label'] = df['final_text'].apply(label_sentiment)
```
