```python
df = pd.read_csv(file_path)

input_col = 'full_text'
df.head()
```

```python
import re

def remove_special_characters(text):
    text = re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", str(text))
    text = re.sub(r"\d+", "", text)
    return text

df['cleaned_text'] = df[input_col].apply(remove_special_characters)
df[['cleaned_text']].head()
```

```python
def case_folding(text):
    return text.lower()

df['case_folded_text'] = df['cleaned_text'].apply(case_folding)
df[['case_folded_text']].head()
```

```python
from nltk.tokenize import word_tokenize
import nltk

# Download necessary NLTK resources
nltk.download('punkt')

# Function to tokenize text
def tokenize_text(text):
    return word_tokenize(text)

# Assuming df and 'case_folded_text' column exist
# Replace df['case_folded_text'] with the actual DataFrame and column name
df['tokenized_text'] = df['case_folded_text'].apply(tokenize_text)
print(df[['tokenized_text']].head())
```
